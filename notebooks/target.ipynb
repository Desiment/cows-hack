```python
columns_to_rewrite = ['id', 'lactation_number', 'birth_difficult', 'days_no_milking', 'days_pregnant', 'group_id', 'prev_group_id','days_milking','date','event','remark', 'age']
df_mastit = pd.DataFrame()
```

```python
%%time
df_mastit = pd.DataFrame()

num_of_cols = len(df.columns)

index_of_col = {}
k = 0
for col in df.columns:
    index_of_col[col] = k
    k += 1
  
    
def write_to_new_row(df, row):
    for col in df.columns:
        if ((col not in columns_to_rewrite) & (row[col] == 0) & (row[col] == 1)):
            new_row[index_of_col[col]] = row[col]
        elif (col in columns_to_rewrite):        
            new_row[index_of_col[col]] = row[col]
            

def write_to_new_row_new_disease(new_row, row):
    if (row.event in diseases.keys()):
        new_row[num_of_cols + diseases[row['event']]] = 1
    
diseases = {'PARES': 0, 'KETOS': 1, 'METR': 2, 'LAME': 3}


def numpyarr_to_dict_for_dfmastit(arr):
    new_dict = {}
    for col in df.columns:
        new_dict[col] = arr[index_of_col[col]]
    for col in diseases:
        new_dict[col] = arr[num_of_cols + diseases[col]]
    new_dict['health'] = arr[num_of_cols + len(diseases)]
    new_dict['protocol_number'] = arr[num_of_cols + len(diseases) + 1]
    return new_dict

def row_to_dfmastit(new_dict):
    global df_mastit
    df_tmp = pd.DataFrame()
    df_tmp = df_tmp.append(new_dict, ignore_index=True)
    df_mastit = pd.concat([df_mastit, df_tmp], ignore_index=True)
    print(df_mastit.shape[0])

def number_of_protocol(MAST_protocol):
    for c in (MAST_protocol):
        if c.isdigit():
            return c
    
i = 0
while(i < 50000):
    row = df.iloc[i]
    idd = row['id']
    new_row = [0] * (num_of_cols + len(diseases) + 1 + 1)
    while((row['id'] == idd) and (row['event'] != 'MAST')):
        write_to_new_row_new_disease(new_row, row)
        write_to_new_row(df, row)
        i += 1        
        row = df.iloc[i]
    if (row['id'] != idd): # корова не болела маститом, перешел к другой корове, не добавляю ничего
        continue
    else:
        MAST_protocol = row['remark']
        index_of_underscore = -1
        for l,c in enumerate(MAST_protocol):
            if (c == '_'):
                index_of_underscore = l
                break
        idd = row['id']
        write_to_new_row(df, row)
        i += 1
        row = df.iloc[i]
        while((row['id'] == idd) & (row['event'] != 'MAST')  & ((row['event'] != 'HEALTH') & ((row['remark'][:4] != 'МАСТ') | (row['remark'][:4] != 'МАТС')))):
            if ((row['event'] == 'MOVE') & ((row['remark'][:4] == 'F003') | (row['remark'][:4] == 'F043'))): #вылечилась
                write_to_new_row(df, row)
                new_row[num_of_cols + len(diseases)] = 1
                new_row[num_of_cols + len(diseases) + 1] = number_of_protocol(MAST_protocol)
                row_to_dfmastit(numpyarr_to_dict_for_dfmastit(new_row))
                break
            write_to_new_row_new_disease(new_row, row)
            write_to_new_row(df, row)
            i += 1
            row = df.iloc[i]
        if (row['id'] != idd): #непонятен результат(нет них health, ни mast, ни move, считаю что не выздоровела)
            new_row[num_of_cols + len(diseases)] = 0
            new_row[num_of_cols + len(diseases) + 1] = number_of_protocol(MAST_protocol)
            row_to_dfmastit(numpyarr_to_dict_for_dfmastit(new_row))
        if ((row['event'] == 'MAST') & (row['remark'] != 'МВ') & (row['remark'] != 'MB')):
            MAST_protocol_new = row['remark']
            index_of_underscore_new = -1
            for l,c in enumerate(MAST_protocol_new):
                if (c == '_'):
                    index_of_underscore_new = l
                    break
            if ((MAST_protocol[index_of_underscore:] not in MAST_protocol_new[index_of_underscore_new:].replace(',', '')) & ('?' not in MAST_protocol_new)):
                write_to_new_row(df, row)
                new_row[num_of_cols + len(diseases)] = 1
                new_row[num_of_cols + len(diseases) + 1] = number_of_protocol(MAST_protocol)
                row_to_dfmastit(numpyarr_to_dict_for_dfmastit(new_row))
            elif (MAST_protocol[index_of_underscore:] in MAST_protocol_new[index_of_underscore_new:].replace(',', '')):
                     new_row[num_of_cols + len(diseases)] = 0
                     new_row[num_of_cols + len(diseases) + 1] = number_of_protocol(MAST_protocol)
                     row_to_dfmastit(numpyarr_to_dict_for_dfmastit(new_row))         
        if((row['event'] == 'HEALTH') & ((row['remark'] == 'МАСТИТ') | (row['remark'] == 'МАТСТИТ') | (row['remark'] == 'МАСТИТ') | (row['remark'] == 'МАСТИТ3'))):
                     new_row[num_of_cols + len(diseases)] = 1
                     new_row[num_of_cols + len(diseases) + 1] = number_of_protocol(MAST_protocol)
                     row_to_dfmastit(numpyarr_to_dict_for_dfmastit(new_row))       
        i += 1        
            
            
            
            
            
        
        
            
```
